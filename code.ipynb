# Data splitting

```python
split_ratio = 0.8
split_idx = int(len(data) * split_ratio)

training_data = data[['Close']].iloc[:split_idx].copy()
testing_data = data[['Close']].iloc[split_idx:].copy()

print(f"Training data shape: {training_data.shape}")
print(f"Testing data shape : {testing_data.shape}")
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0, 1))
```

```python
training_data_array = scaler.fit_transform(training_data)
training_data_array
```

```python
training_data_array.shape[0]
```

```python
x_train = []
y_train = []

for i in range(100, training_data_array.shape[0]):
    x_train.append(training_data_array[i-100:i])
    y_train.append(training_data_array[i, 0])

x_train, y_train  = np.array(x_train), np.array(y_train)
```

```python
print("x_train shape:",x_train.shape)
print("y_train shape:",y_train.shape)
```

```python
# Model Building
from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential
```

```python
model = Sequential()

model.add(LSTM(units = 50, activation = 'relu', return_sequences = True, input_shape = (x_train.shape[1],1)))
model.add(Dropout(0.2))

model.add(LSTM(units = 60, activation = 'relu', return_sequences = True))
model.add(Dropout(0.3))

model.add(LSTM(units = 80, activation = 'relu', return_sequences = True))
model.add(Dropout(0.4))

model.add(LSTM(units = 120, activation = 'relu'))
model.add(Dropout(0.5))

model.add(Dense(units = 1))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
model.fit(x_train, y_train, epochs = 50)
```

```python
past_100_days = training_data.tail(100)
```

```python
final_df = pd.concat([past_100_days, testing_data], ignore_index=True)
```

```python
final_df
```

```python
input_data = scaler.fit_transform(final_df)
```

```python
x_test = []
y_test = []

for i in range(100, input_data.shape[0]):
    x_test.append(input_data[i-100:i])
    y_test.append(input_data[i, 0])

x_test, y_test  = np.array(x_test), np.array(y_test)
```

```python
x_test.shape
```

```python
y_predicted = model.predict(x_test)
```

```python

y_predicted.shape
```

```python
scaler.scale_
```

```python
scaler_factor = 1 / 0.0035166
y_predicted = y_predicted * scaler_factor
y_test = y_test * scaler_factor
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y_test, label = 'Original Price', linewidth = 1)
plt.plot(y_predicted, label = 'Predicted Price', linewidth = 1)
plt.legend()
plt.show()
```

```python
y_predicted_original = scaler.inverse_transform(y_predicted)
y_test_original = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test_original, y_predicted_original)
mae = mean_absolute_error(y_test_original, y_predicted_original)
r2 = r2_score(y_test_original, y_predicted_original)

print(f"MSE: {mse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"R² : {r2:.4f}")
```

```python
from sklearn.metrics import mean_absolute_percentage_error

mape = mean_absolute_percentage_error(y_test_original, y_predicted_original)
print(f"Mean Absolute Percentage Error (MAPE): {mape * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(y_test_original, label='Actual', linewidth = 1)
plt.plot(y_predicted_original, label='Predicted',  linewidth = 1)
plt.title('Actual vs Predicted Values')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```
